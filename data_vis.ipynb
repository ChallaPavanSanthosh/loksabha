```python
votes_by_state = df.groupby('State')['Total Votes'].sum().sort_values(ascending=False)

# Plotting
plt.figure(figsize=(12, 6))
votes_by_state.plot(kind='bar', color='skyblue')
plt.title('Total Votes by State')
plt.xlabel('State')
plt.ylabel('Total Votes')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Step 1: Counting Occurrences by Party
party_counts = df['Party'].value_counts().reset_index()
party_counts.columns = ['Party', 'Count']

# Step 2: Identifying Top 10 Parties
top_10_parties = party_counts.head(10)

# Step 3: Summarizing Other Parties
other_count = party_counts.iloc[10:].sum()
other_summary = pd.DataFrame({'Party': ['Others'], 'Count': [other_count['Count']]})

# Step 4: Combining Top 10 and Others
final_counts = pd.concat([top_10_parties, other_summary])

# Step 5: Plotting a Bar Graph
plt.figure(figsize=(10, 6))
bars = plt.bar(final_counts['Party'], final_counts['Count'], color='blue')

# Labeling the "Others" category
bars[-1].set_label('Others')

plt.title('Number Of Seats Won By Each Party')
plt.xlabel('Party Names')
plt.ylabel('Total Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.legend()
plt.show()
```

```python
rahul_entries = df[df['Winning Candidate'] == 'RAHUL GANDHI']
modi_entries = df[df['Winning Candidate'] == 'NARENDRA MODI']
amit_entries = df[df['Winning Candidate'] == 'AMIT SHAH']

# Get the votes for Rahul Gandhi, Narendra Modi, and Amit Shah
rahul_votes = rahul_entries['Margin'].values
modi_votes = modi_entries['Margin'].values[0] if not modi_entries.empty else 0
amit_votes = amit_entries['Margin'].values[0] if not amit_entries.empty else 0

# Get the original constituency names for Rahul Gandhi
rahul_constituencies = list(rahul_entries['Parliament Constituency'])

# Get the original constituency name for Narendra Modi
modi_constituency = modi_entries['Parliament Constituency'].values[0] if not modi_entries.empty else "Modi Constituency"

# Get the original constituency name for Amit Shah
amit_constituency = amit_entries['Parliament Constituency'].values[0] if not amit_entries.empty else "Amit Shah Constituency"

# Combine the data
data_to_plot = pd.DataFrame({
    'Candidate': ['Rahul Gandhi'] * len(rahul_votes) + ['Narendra Modi', 'Amit Shah'],
    'Constituency': rahul_constituencies + [modi_constituency, amit_constituency],
    'Votes': list(rahul_votes) + [modi_votes, amit_votes]
})

# Plot the comparison
plt.figure(figsize=(12, 6))
sns.barplot(data=data_to_plot, x='Constituency', y='Votes', hue='Candidate', palette='muted')
plt.title('Comparison of Votes for Rahul Gandhi, Narendra Modi, and Amit Shah')
plt.xlabel('Constituency')
plt.ylabel('Votes')
plt.xticks(rotation=45)
plt.show()
```

```python
df['Total Votes'] = pd.to_numeric(df['Total Votes'], errors='coerce')

# Drop rows with NaN values in 'Total Votes'
df = df.dropna(subset=['Total Votes'])

# Convert 'Total Votes' to integer
df['Total Votes'] = df['Total Votes'].astype(int)

# Aggregating total votes by party
aggregated_data = df.groupby('Party')['Total Votes'].sum()

# Sorting the aggregated data to get the top 9 parties
top_parties = aggregated_data.sort_values(ascending=False).head(9)

# Summing the votes for other parties
others = aggregated_data.sort_values(ascending=False).iloc[9:].sum()

# Adding the 'Others' category to the top parties
top_parties['Others'] = others

# Pie chart: Vote-wise Share for Top 9 Parties and Others
plt.figure(figsize=(10, 10))
top_parties.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=sns.color_palette("Set3"))
plt.title('Vote Wise Share For Each Party')
plt.ylabel('')
plt.show()
```

```python
# Count the number of seats won by each state
state_seat_counts = df['State'].value_counts()

# Plot the pie chart
plt.figure(figsize=(10, 8))
plt.pie(state_seat_counts, labels=state_seat_counts.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.tab20.colors)
plt.title('State wise Seats')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.legend(loc='upper right', bbox_to_anchor=(1.3, 0.9), title="States")
plt.show()
```

```python

from matplotlib import pyplot as plt
import seaborn as sns
def _plot_series(series, series_name, series_index=0):
  palette = list(sns.palettes.mpl_palette('Dark2'))
  xs = series['Winning Candidate']
  ys = series['Margin']

  plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])

fig, ax = plt.subplots(figsize=(10, 5.2), layout='constrained')
df_sorted = df.sort_values('Winning Candidate', ascending=True)
_plot_series(df_sorted, '')
sns.despine(fig=fig, ax=ax)
plt.xlabel('Winning Candidate')
_ = plt.ylabel('Margin')
```

```python
# Group data by Alliance and count the occurrences of each Parliament_Constituency
alliance_counts = df.groupby('Alliance').count()['Parliament Constituency']

# Plotting
fig, ax = plt.subplots()

# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = alliance_counts.index
sizes = alliance_counts.values
colors = plt.cm.Paired(range(len(labels)))

# Create a pie chart
ax.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=colors, wedgeprops=dict(width=0.3))

# Draw circle for the donut chart
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig.gca().add_artist(centre_circle)

# Equal aspect ratio ensures that pie is drawn as a circle.
ax.axis('equal')

plt.title('Parliament Constituencies by Alliance')
plt.show()
```
